```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
# import plotly.io as pio
# pio.renderers.default = 'browser'  # or 'notebook', 'svg', etc.

# Load the Netflix dataset (update the path as necessary)
df = pd.read_csv('netflix_titles.csv')  # Update this path if necessary

# Display the first few rows of the dataframe
print(df.head())

# Check for any missing values in the date_added column
print("Missing values in date_added:", df['date_added'].isnull().sum())  # Count missing values

# Drop rows with missing date_added values
df = df.dropna(subset=['date_added'])

# Clean date_added: Remove any leading/trailing whitespace
df['date_added'] = df['date_added'].str.strip()

# Convert date_added to datetime format with error handling
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce', infer_datetime_format=True)

# Check if any NaT values remain after conversion
print("NaT values in date_added after conversion:", df['date_added'].isnull().sum())  # Count NaT values

# Extract year from the date added column
df['year_added'] = df['date_added'].dt.year

# Data Visualization with Matplotlib

# Count the number of Movies and TV Shows
content_counts = df['type'].value_counts()

# Bar Chart: Count of Movies vs. TV Shows
plt.figure(figsize=(8, 5))
plt.bar(content_counts.index, content_counts.values, color=['blue', 'orange'])
plt.title('Count of Movies vs. TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
# Count Plot: Count of Movies vs. TV Shows
# plt.figure(figsize=(8, 5))
# sns.countplot(x='type', data=df, palette='Set2', order=df['type'].value_counts().index)
# plt.title('Count of Movies vs. TV Shows')
# plt.xlabel('Type')
# plt.ylabel('Count')
# plt.show()

# Clean the duration column
df['duration'] = df['duration'].str.replace(' min', '', regex=False)  # Remove ' min'
df['duration'] = pd.to_numeric(df['duration'], errors='coerce')  # Convert to numeric, coercing errors

# Print unique values in duration after conversion to check for issues
print("Unique values in duration column after cleaning:")
print(df['duration'].unique())

# Drop rows with NaN values in duration
df = df.dropna(subset=['duration'])

# Histogram: Plot a Histogram of Movie Durations
plt.figure(figsize=(10, 6))
plt.hist(df['duration'].dropna(), bins=30, color='purple', alpha=0.7)
plt.title('Histogram of Movie Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()

# Count of content added per year
content_trend = df['year_added'].value_counts().sort_index().reset_index()
content_trend.columns = ['year_added', 'count']  # Rename columns for clarity

# Interactive Line Chart: Trend of Content Added Over the Years
fig = px.line(content_trend, 
              x='year_added', 
              y='count', 
              title='Trend of Content Added Over the Years',
              labels={'year_added': 'Year', 'count': 'Count of Content Added'})
fig.show()

# Data Visualization with Seaborn
df = pd.read_csv('netflix_titles.csv')  # Update this path if necessary

# Display unique values in the type column
print("Unique content types:", df['type'].unique())

# Display counts of each content type
print("Count of each type:\n", df['type'].value_counts())

# If TV Shows are present, create the count plot
if 'TV Show' in df['type'].unique():
    plt.figure(figsize=(8, 5))
    sns.countplot(x='type', data=df, palette='Set2')
    plt.title('Count of Movies vs. TV Shows')
    plt.xlabel('Type')
    plt.ylabel('Count')
    plt.show()
else:
    print("No TV Shows found in the dataset.")

# Box Plot: Distribution of Movie Durations
plt.figure(figsize=(10, 6))
sns.boxplot(x='type', y='duration', data=df[df['type'] == 'Movie'], palette='Set3')
plt.title('Distribution of Movie Durations')
plt.xlabel('Type')
plt.ylabel('Duration (minutes)')
plt.show()


# Check for non-numeric columns before creating the correlation heatmap
non_numeric_columns = df.select_dtypes(exclude='number').columns
print("Non-numeric columns in DataFrame:", non_numeric_columns)

# Convert non-numeric columns to numeric if possible
for col in non_numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Create a correlation heatmap
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

# Interactive Visualizations with Plotly

# Scatter Plot: Interactive Scatter Plot of Release Year vs. Duration for Movies
df = pd.read_csv('netflix_titles.csv') 

# Clean column names to remove any leading/trailing whitespace
df.columns = df.columns.str.strip()

# Filter for Movies
movies_df = df[df['type'] == 'Movie']
print("Shape of movies DataFrame:", movies_df.shape)

# Display the first few rows to inspect relevant columns
print(movies_df[['release_year', 'duration']].head(10))

# Clean the duration column
# Remove ' min' and convert to numeric
movies_df['duration'] = movies_df['duration'].str.replace(' min', '', regex=False)  # Remove ' min'
movies_df['duration'] = pd.to_numeric(movies_df['duration'], errors='coerce')  # Convert to numeric, coercing errors

# Check for NaN values in duration after conversion
print("Missing values in duration column after conversion:", movies_df['duration'].isnull().sum())

# Drop rows with NaN values in duration
movies_df = movies_df.dropna(subset=['duration'])
print("Shape of movies DataFrame after cleaning duration:", movies_df.shape)

# Scatter Plot: Interactive Scatter Plot of Release Year vs. Duration for Movies
fig = px.scatter(movies_df,
                 x='release_year',
                 y='duration',
                 title='Release Year vs. Duration for Movies',
                 labels={'release_year': 'Release Year', 'duration': 'Duration (minutes)'})
fig.show()

# Pie Chart: Distribution of Content Ratings
df = pd.read_csv('netflix_titles.csv')  # Update this path if necessary

# Display the first few rows of the dataframe
print("Original DataFrame shape:", df.shape)
print("First few rows of the DataFrame:\n", df.head(10))

# Strip whitespace from the rating column
df['rating'] = df['rating'].str.strip()

# Drop rows with missing ratings
df = df.dropna(subset=['rating'])
print("Shape after dropping missing ratings:", df.shape)

# Create a DataFrame with counts of each rating
rating_counts = df['rating'].value_counts().reset_index()
rating_counts.columns = ['rating', 'count']  # Rename columns for clarity
print("Rating counts after cleaning:\n", rating_counts)

# Create the pie chart
fig = px.pie(rating_counts, 
             names='rating', 
             values='count', 
             title='Distribution of Content Ratings', 
             color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()

# Interactive Visualizations: Trend of Content Added Over the Years
fig = px.line(content_trend.reset_index(), 
              x='index', 
              y='year_added', 
              title='Trend of Content Added Over the Years',
              labels={'index': 'Year', 'year_added': 'Count of Content Added'})
fig.show()
```
